```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Load the data
fct = pd.read_csv('FCT-ABDUCTION.csv', encoding='latin-1')

# Check size
print(f"Rows: {fct.shape[0]}, Columns: {fct.shape[1]}")

# Check column names
print(fct.columns.tolist())

# Check data types
print(fct.dtypes)

# First few rows
fct.head()
```

```python
# Check for missing values
fct.isnull().sum()
```

```python
# Look at the last 3 rows
fct.tail(3)
```

```python
# Drop the last row
fct = fct.drop(fct.index[-1])

# Confirm removal
fct.tail()
```

```python
# Recheck for missing values
fct.isnull().sum()
```

```python
# Check for duplicates
print(f"Number of duplicate rows: {fct.duplicated().sum()}")
```

```python
# Show the duplicate rows so you can see what they are
fct[fct.duplicated(keep=False)]
```

```python
# Remove duplicates, keeping the first occurrence
fct = fct.drop_duplicates()
```

```python
# Check duplicate count again
print(f"Remaining duplicates: {fct.duplicated().sum()}")
```

```python
# Convert to datetime
fct['DATE'] = pd.to_datetime(fct['DATE'])

# Now you can extract useful components
fct['YEAR'] = fct['DATE'].dt.year
fct['MONTH'] = fct['DATE'].dt.month
fct['DAY'] = fct['DATE'].dt.day
fct['DAY_OF_WEEK'] = fct['DATE'].dt.day_name()
```

```python
# Check the data types
print(fct[['NUMBER OF PEOPLE KIDNAPPED', 'NUMBER OF PEOPLE KILLED', 'NUMBER OF PEOPLE INJURED']].dtypes)

# Check for any negative values or impossibly large values
print(fct[['NUMBER OF PEOPLE KIDNAPPED', 'NUMBER OF PEOPLE KILLED', 'NUMBER OF PEOPLE INJURED']].describe())
```

```python
# Check unique LGA values
fct['LAC'].unique()
```

```python
# Check unique threat actor values
fct['THREAT ACTORS'].unique()
```

```python
# Standardize threat actor categories (shorten long names only)
fct['THREAT ACTORS'] = fct['THREAT ACTORS'].replace({
    'NSAG - Non State Armed Group': 'NSAG',
    'Security Forces (Military, Police, Paramilitary, DSS / SSS and including MNJTF)': 'Security Forces',
    'Political Thug/Hired Assailant': 'Political Thug'
})
```

```python
fct['THREAT ACTORS'].value_counts()
```

```python
# Show all column names
fct.columns.tolist()
```

```python
# Rename the column
fct.rename(columns={'AFFAECTED GROUOS': 'AFFECTED GROUPS'}, inplace=True)

# Verify the change
print(fct.columns.tolist())
```

```python
# Check rows where THREAT ACTORS is 'Civilians'
fct[fct['THREAT ACTORS'] == 'Civilians'][['THREAT ACTORS', 'AFFECTED GROUPS', 'BREAKDOWN OF INCIDENT']]
```

```python
# Incidents by year
fct['YEAR'].value_counts().sort_index()
```

```python
# Incidents by LGA
fct['LAC'].value_counts()
```

```python
# Threat actor distribution
fct['THREAT ACTORS'].value_counts()
```

```python
# Check if 2025 is complete
fct[fct['YEAR'] == 2025]['MONTH'].value_counts().sort_index()
```

```python
# Total kidnapped by LGA
fct.groupby('LAC')['NUMBER OF PEOPLE KIDNAPPED'].sum().sort_values(ascending=False)
```

```python
# Incidents by month (all years combined)
fct['MONTH'].value_counts().sort_index()
```

```python
# Average killed by threat actor
fct.groupby('THREAT ACTORS')['NUMBER OF PEOPLE KILLED'].mean().sort_values(ascending=False)
```

```python
# Scatter plot or correlation
fct[['NUMBER OF PEOPLE KIDNAPPED', 'NUMBER OF PEOPLE KILLED']].corr()
```

```python
# Check for ransom mentions
fct[fct['BREAKDOWN OF INCIDENT'].str.contains('ransom', case=False, na=False)][['BREAKDOWN OF INCIDENT', 'THREAT ACTORS']].head(10)
```

```python
# Check for deaths
fct[fct['NUMBER OF PEOPLE KILLED'] > 0][['BREAKDOWN OF INCIDENT', 'NUMBER OF PEOPLE KILLED', 'THREAT ACTORS']].head(10)
```

```python
# Check for large-scale kidnappings (10+ people)
fct[fct['NUMBER OF PEOPLE KIDNAPPED'] >= 10][['BREAKDOWN OF INCIDENT', 'NUMBER OF PEOPLE KIDNAPPED', 'LAC']].head(10)
```

```python
# Check for mentions of "ransom"
fct[fct['BREAKDOWN OF INCIDENT'].str.contains('ransom', case=False, na=False)]['THREAT ACTORS'].value_counts()
```

```python
# Check for mentions of "security" or "police"
fct[fct['BREAKDOWN OF INCIDENT'].str.contains('police|security|military|army', case=False, na=False)]['THREAT ACTORS'].value_counts()
```

```python
# Check for mentions of "school" or "student"
fct[fct['BREAKDOWN OF INCIDENT'].str.contains('school|student|teacher', case=False, na=False)][['BREAKDOWN OF INCIDENT', 'LAC']]
```

```python
# Check for mentions of "women" or "children"
fct[fct['BREAKDOWN OF INCIDENT'].str.contains('woman|women|child|children', case=False, na=False)]['LAC'].value_counts()
```

```python
# Save the cleaned dataframe to CSV
fct.to_csv('fct_abduction_cleaned.csv', index=False)

print("File saved successfully!")
```
